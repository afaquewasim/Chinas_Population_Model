```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/Chinas Population En.csv')
df.head()
```

```python
df = df.rename(columns={'Unnamed: 0': 'Year'})
```

```python
df.head()
```

```python
plt.xlabel("Female population")
plt.ylabel("Urban population")

plt.plot(df['Female population (ten thousand people)'],df['Urban population (ten thousand people)'])
plt.show()

```

```python
plt.xlabel("Female population")
plt.ylabel("total population")
plt.subplot(2, 1, 1)
plt.plot(df['Female population (ten thousand people)'],df['At the end of the total population (ten thousand people)'])
plt.show()


plt.xlabel("male population")
plt.ylabel("total population")
plt.subplot(2, 1, 1)
plt.plot(df['The male population (ten thousand people)'],df['At the end of the total population (ten thousand people)'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('Urban population (ten thousand people)', axis=1)
y = df['Urban population (ten thousand people)']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) 
```

```python
from sklearn.linear_model import LinearRegression

Model = LinearRegression()

Model.fit(x_train, y_train)

Model.score(x_test, y_test)
```
